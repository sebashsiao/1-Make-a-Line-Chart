Exercise 6: Add title and legend

#### Setup

```python
# import matplotlib and pandas
from matplotlib import pyplot as plt
import pandas as pd

# load dataset
line_data = pd.read_csv('2020-monthly-avg-temps-f.csv')
```

```python
# display the graphs correctly
%matplotlib inline
plt.rcParams['figure.figsize'] = (5, 3)
plt.rcParams['figure.dpi'] = 75
```

#### 1. Title the graph “Average High Temps Around the World, 2020”.

```python
## YOUR SOLUTION HERE ##
plt.plot(line_data.month_name, line_data.alice_springs_avg_high, color='green', linestyle='dotted')
plt.plot(line_data.month_name, line_data.windhoek_avg_high, color='green', linestyle='dashed')
plt.plot(line_data.month_name, line_data.quito_avg_high, color='steelblue', linewidth=2.5)
plt.plot(line_data.month_name, line_data.samarqand_avg_high, color='darkorange', linestyle='dotted')
plt.plot(line_data.month_name, line_data.london_avg_high, color='darkorange', linestyle='dashed')
plt.plot(line_data.month_name, line_data.kodiak_avg_high, color='darkorange', linestyle='solid')
plt.title('Average High Temps Around the World, 2020')

## show outputs 
plt.show()
```

#### 2. Add the city name as a legend label to each line and create the legend.

```python
## YOUR SOLUTION HERE ##
plt.plot(line_data.month_name, line_data.alice_springs_avg_high, color='green', linestyle='dotted', label='Alice Springs')
plt.plot(line_data.month_name, line_data.windhoek_avg_high, color='green', linestyle='dashed', label='Windhoek')
plt.plot(line_data.month_name, line_data.quito_avg_high, color='steelblue', linewidth=2.5, label='Quito')
plt.plot(line_data.month_name, line_data.samarqand_avg_high, color='darkorange', linestyle='dotted', label='Samarqand')
plt.plot(line_data.month_name, line_data.london_avg_high, color='darkorange', linestyle='dashed', label='London')
plt.plot(line_data.month_name, line_data.kodiak_avg_high, color='darkorange', linestyle='solid', label='Kodiak')
plt.title('Average High Temps Around the World, 2020')

## show outputs 
plt.show()
```
